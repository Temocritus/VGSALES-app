```python
#Data Processing
```

```python
# 1. Finding the leading gaming platforms wrt sales, both globally and regionally
```

```python
       # i.finding the leading console globally
    
df3 = df.groupby(['Platform']).agg({'Global_Sales': 'sum'}).reset_index()
df3
```

```python
  #rearranging the global sales in descending order 
    
df4 = df3.sort_values('Global_Sales',ascending = False)
df4
```

```python
   #Getting top 10 leading platforms 

Global=df4.head(10)
Global
```

```python
#Plotting  the bar chart of popular sales wrt sales

Global.plot.bar(x="Platform", y="Global_Sales", rot=70, title="Sales by Gaming Platform Globally, Top 10");

plt.show(block=True)
```

```python
  # ii. Deducing the leading platforms in the North America region
    
df5 = df.groupby(['Platform']).agg({'NA_Sales': 'sum'}).reset_index()
df5
```

```python
#descending order

America = df5.sort_values('NA_Sales',ascending = False)
America
```

```python
NA = America.head(10)
NA
```

```python
# VIsualising
NA.plot.bar(x="Platform", y="NA_Sales", rot=70, title="Game Sales in North America, Top 10 Platforms");

plt.show(block=True)
```

```python
  # European Region
    
Europe = df.groupby(['Platform']).agg({'EU_Sales': 'sum'}).reset_index()
Europe
```

```python
EU1 = Europe.sort_values('EU_Sales',ascending = False)
EU1
```

```python
EU = EU1.head(10)
EU
```

```python
# VIsualising

EU.plot.bar(x="Platform", y="EU_Sales", rot=70, title="Game Sales in North America, Top 10 Platforms");

plt.show(block=True)
```

```python
  # iii. Japan Region
    
Japan= df.groupby(['Platform']).agg({'JP_Sales': 'sum'}).reset_index()
Japan    
    
```

```python
JP1 = Japan.sort_values('JP_Sales',ascending = False)
```

```python
JP = JP1.head(10)
JP
```

```python
# VIsualising

JP.plot.bar(x="Platform", y="JP_Sales", rot=70, title="Game Sales in Japan, Top 10 Platforms");

plt.show(block=True)
```

```python
  # Rest of the world
Rest = df.groupby(['Platform']).agg({'Other_Sales': 'sum'}).reset_index()
Rest   
```

```python
Other1 = Rest.sort_values('Other_Sales',ascending = False)
```

```python
Other = Other1.head(10)
Other
```

```python
# VIsualising

Other.plot.bar(x="Platform", y="Other_Sales", rot=70, title="Game Sales for the rest of the world, Top 10 Platforms");

plt.show(block=True)
```

```python
   # combining the top 10 leading Platforms' data frames
    
World = pd.concat([Global, NA, EU, JP, Other], axis=1)
World
    
    
```

```python
  #dropping columns with similr names
    
World_sales=World.iloc[:,~World.columns.duplicated()]
World_sales    
```

```python
    # arranging the data frame in descending order wrt Global sales

Platforms= World_sales.sort_values('Global_Sales',ascending = False)
Platforms
```

```python
TopPlatforms = Platforms.head(10)
TopPlatforms
```

```python
  #Global sales dictate popularity of the gaming platform, thus platforms that are globally popular but not regionally popular are assigned a zero for the corresponding region
    
Popular_Consoles= TopPlatforms.replace(to_replace = np.nan, value =0)
Popular_Consoles    
```

```python
#Plotting a line chart for all sales

plt.title('Top 10 popular platforms with respect to Game Sales')

plt.plot(Popular_Consoles['Platform'], Popular_Consoles['NA_Sales'], label='NA Sales')

plt.plot(Popular_Consoles['Platform'], Popular_Consoles['EU_Sales'], label='EU Sales')

plt.plot(Popular_Consoles['Platform'], Popular_Consoles['JP_Sales'], label='JP Sales')

plt.plot(Popular_Consoles['Platform'], Popular_Consoles['Other_Sales'], label ='Other Sales')

plt.plot(Popular_Consoles['Platform'], Popular_Consoles['Global_Sales'], label ='Global Sales')

plt.legend()

plt.show()
```

```python
  
 # Popular publisher with respect to number of games in the list
```

```python
# Determinig entries made by publisher. This informs on the numner of games per publisher
# transforming value counts to a dataframe

Publishers = (df['Publisher'].value_counts().rename_axis('unique_values').to_frame('Counts').reset_index())
Publishers
```

```python

Top_publishers = Publishers.nlargest(10, 'Counts')
```

```python
#renaming a colomn header

Popular_Publishers = Top_publishers.rename(columns={'unique_values': 'Publishers'})
Popular_Publishers
```

```python
# Visualisation into a table
HTML(Popular_Publishers.to_html( classes= 'table table-striped table-bordered' ))
```

```python
# Visualisation of game enties per Publisher

Popular_Publishers.plot.bar(x="Publishers", y="Counts", rot=70, title="Leading Publishers by game entries");

plt.show(block=True)
```
